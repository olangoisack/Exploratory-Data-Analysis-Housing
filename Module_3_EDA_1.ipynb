```python
#Descriptive statistics are used to summarize the main features of data
#Mean - average value
#Standard Deviation(std) - how spread out the values are from the mean*
#Median - middle value
```

```python
import pandas as pd 

df = pd.DataFrame({"Age": [18, 22, 25, 29, 34, 40, 50]})
print(df["Age"].mean()) #mean
                           
```

```python
print(df["Age"].median()) 
```

```python
print(df["Age"].std()) 
```

```python
print(df.describe()) # quick summary
```

# Kernel Density Plot(KDE)

```python
import warnings
warnings.filterwarnings('ignore')

sns.kdeplot(df["Age"], fill=True)
plt.title("KDE of Age")
plt.show()
```

```python
#iInterpretation of this plot(KDE plot)
#peak -mode
#Width = spread/variance
#Asymetry = skewness
```
