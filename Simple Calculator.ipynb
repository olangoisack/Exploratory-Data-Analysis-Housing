# Simple Calculator in Python

```python
#Step 1:Ask the user for the first number

num1 = float(input("Enter first number: "))
```

```python
#Step 2:Ask the user for the operator(+, -, *, /)

operator = input("Enter operator (+, -, *, /): ")
```

```python
#Step 3:Ask the user for the second number
num2 = float(input("Enter second number: "))
```

```python
##### Step 4:Check which operator the user entered and perform the calculation

if operator == "+":                  #If user entered '+', perform addition
    print("Result:", num1 + num2)   
elif operator == "-":                #If user entered '-',perform subtraction
    print("Result:", num1 - num2)
elif operator == "*":                 #If user entered '*',perform multiplication
    print("Result:", num1 * num2)    
elif operator == "/":                #If user entered '/',perform division
    if num2 != 0:                 #Prevents division by zero
        print("Result", num1 / num2)
    else:
        print("Erro:Division by zero")
else:
    #If user entered an operator that is not +,-,*,/
    print("Invalid operator")
        
```
