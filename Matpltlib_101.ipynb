```python
import matplotlib.pyplot as plt
```

# Basic Plot

```python
x = [1,2,3,4,5] # a list of integers -this will appear on the horizontal
y = [2,4,6,8,10] # a list of even numbers -this will appear on the vertical

plt.plot(x,y)
plt.title("Basic Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

# Different Types of Plots

## 1.Line Plot

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(x,y, color='#e38914',marker='x')
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

# 2.Bar Chart

```python
students = ["Alice","Bob","Charlie","Daudi","Eve"]
marks = [85,90,78,88,95]

plt.bar(students,marks,color='green')
plt.title("Bar Chart-Student Marks")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

# 3.Scatter Plot

```python
x = [5,7,8,7,6,9,5,6,7,8]
y = [99,86,87,88,100,86,103,87,94,78]

plt.scatter(x,y,color='red')
plt.title("Scatter Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

# 4.Histogram

```python
import numpy as np

data = np.random.randn(1000)  #1000 random numbers

plt.hist(data,bins=20,color='purple',edgecolor='black')
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frquency")
plt.show()
```

# 5.Pie Chart

```python
activities = ["Study","Sleep","Play","Eat"]
hours = [7,8,5,4]

plt.pie(hours,labels=activities,autopct='%1.1f%%',startangle=90)
plt.title('Pie Chart -Daily Activities')
plt.show()
```
