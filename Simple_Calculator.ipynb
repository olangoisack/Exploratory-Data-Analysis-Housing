# Simple Calculator in Python

```python
#Step 1:Ask user for the first number
num1 = float(input("Enter first number:")) 

#Step 2:Ask user for the operator
operator = input("Enter the operator (+,-,*,/):")

#Step 3:Ask user for the second number
num2 = float(input("Enter second number: "))

#Step 4:Check which operator the user entered and perform the calculation
if operator == "+":              #If user entered '+', perform addition
    print("Result:",num1 + num2)
elif operator == "-":            #If user entered '-',perform subtraction
    print("Result:", num1 - num2)
elif operator == "*":            #If user entered '*',perform multiplication
    print("Result:",num1 * num2)
elif operator == "/":           #If user entered '/' perform division
    if num2 != 0:
        print("Result:", num1 / num2)
    else:
        print("Error:Division by a zero")
else:
    #If the user enters an operator that is not +,-,*,/
    print("Invalid operator!")
    

    
```

# Modified Calculator

```python
#Looping Calculator in Python
#This calculator keeps running until the user chooses to quit

while True: #Step 1:Create an infinite loop to keep the calculator running
    
    #Step 2:Ask user for the first number
    num1 = float(input("Enter first number:")) 

    
    #Step 3:Ask user for the operator(+,-,*,/) or 'q' to quit
    operator = input("Enter the operator (+,-,*,/) 'q' to quit:")

    #Step4:Check if the user wants to quit
    if operator == "q":
        print("Goodbye")  #Exits the message
        break             #Stop the loop and end the program
    #Step 5:Ask user for the second number
    num2 = float(input("Enter second number: "))   

    #Step6:Perform the calculation based on the chosen operator
    if operator == "+":              #If user entered '+', perform addition
        print("Result:",num1 + num2)
    elif operator == "-":            #If user entered '-',perform subtraction
        print("Result:", num1 - num2)
    elif operator == "*":            #If user entered '*',perform multiplication
        print("Result:",num1 * num2)
    elif operator == "/":           #If user entered '/' perform division
        if num2 != 0:
            print("Result:", num1 / num2)
        else:
            print("Error:Division by a zero")
    else:
    #If the user enters an operator that is not +,-,*,/
        print("Invalid operator!")
```
