```python
# Step 1: Import pandas
import pandas as pd

# Step 2: Create some sample data
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "Country": ["Kenya", "USA", "UK"]
}

# Step 3: Turn it into a DataFrame
df = pd.DataFrame(data)
print("This is our DataFrame:")
print(df)

# Step 4: Save into different formats
# Save as CSV
df.to_csv("output.csv", index=False)

# Save as Excel
df.to_excel("output.xlsx", index=False)

# Save as JSON
df.to_json("output.json", orient="records", indent=4)

print("\nFiles saved successfully: output.csv, output.xlsx, output.json")
```
