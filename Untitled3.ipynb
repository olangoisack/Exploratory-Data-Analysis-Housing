```python
# Module 1: Data Wrangling with Pandas
# Objective: Learn how to load, clean, explore, and manipulate datasets using Pandas.

# --- 1. Reading and Writing Data (CSV, Excel, JSON) ---
import pandas as pd

# Reading CSV file
df_csv = pd.read_csv("students.csv")  # Replace with actual file path

# Reading Excel file
df_excel = pd.read_excel("students.xlsx")  # Requires openpyxl installed

# Reading JSON file
df_json = pd.read_json("students.json")

# Writing DataFrame to CSV
df_csv.to_csv("output.csv", index=False)

# --- 2. Pandas DataFrames and Series ---
# Series: one-dimensional labeled array
ages = pd.Series([22, 25, 30], name="Age")
print("Series Example:\n", ages)

# DataFrame: two-dimensional table (rows & columns)
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [22, 25, 30],
    "City": ["Nairobi", "Lagos", "Cairo"]
}
df = pd.DataFrame(data)
print("\nDataFrame Example:\n", df)

# --- 3. Indexing, Filtering, and Slicing ---
# Selecting columns
print(df["Name"])  # Single column
print(df[["Name", "City"]])  # Multiple columns

# Selecting rows
print(df.iloc[0])          # First row by index position
print(df.loc[1, "Age"])   # Row with index=1, column="Age"

# Filtering with conditions
print(df[df["Age"] > 23])
print(df[(df["Age"] > 23) & (df["City"] == "Nairobi")])

# --- 4. Merging and Concatenation ---
# Concatenation example
df1 = pd.DataFrame({"ID": [1, 2], "Name": ["Alice", "Bob"]})
df2 = pd.DataFrame({"ID": [1, 2], "Score": [90, 85]})

concat_df = pd.concat([df1, df2], axis=1)
print("\nConcatenated DataFrame:\n", concat_df)

# Merging example
merged = pd.merge(df1, df2, on="ID")
print("\nMerged DataFrame:\n", merged)

# --- 5. Basic Data Inspection Methods ---
print("\nShape of DataFrame:", df.shape)   # Rows, columns
print("First 5 rows:\n", df.head())
print("Last 5 rows:\n", df.tail())
print("DataFrame info:")
print(df.info())
print("\nSummary statistics:\n", df.describe())
print("\nValue counts for City column:\n", df["City"].value_counts())

# --- Exercises for Students ---
# 1. Load a CSV file and export it to Excel.
# 2. Create your own DataFrame with at least 3 columns.
# 3. Select all students older than 23 who live in Nairobi.
# 4. Merge two datasets on a common key and inspect the result.
# 5. Find how many unique cities are in your dataset and which appears most often.
```
