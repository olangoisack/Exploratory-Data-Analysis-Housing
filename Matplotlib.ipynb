```python
import matplotlib.pyplot as plt
```

# Basic Plot

```python
x  = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(x,y)
plt.title("Basic Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

# Different Plot Types

```python
#Line Plot -plt.plot()
#Bar chart - plt.bar()
#Scatter plot - plt.scatter()
#Histogram - plt.hist()
```

# Styling

```python
#Colours ,line styles,markers
```

```python
plt.plot(x,y,color='red',linestyle='--',marker='o')
```

# Multiple Plots

```python
plt.plot(x,y,label="y=2x")
plt.plot(x,[i**2 for i in x],label="y=x^2")
plt.legend()
plt.show()
```
