# While LOOP

```python
count = 1  #starting point
while count <= 5:                   #keeps looping while count is less than or equak to 5
    print(count)

    count = count + 1  #increase count by 1
```

# For Loop

```python
#Repeat something a fixed number of times.
#For example

for num in range(1,6): #range(1,6) means 1,2,3,4,5

    print(num)
```
