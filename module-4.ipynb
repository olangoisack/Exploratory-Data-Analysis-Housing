```python
import pandas as pd, seaborn as sns, matplotlib.pyplot as plt
sns.set(style="whitegrid")

df = pd.read_csv('/kaggle/input/d/spscientist/students-performance-in-exams/StudentsPerformance.csv')

#Compute average scores
avg_scores = df[['math score','reading score', 'writing score']].mean()
```

# Line Plot

```python
sns.lineplot(x=avg_scores.index, y=avg_scores.values, marker='o')
plt.title("Average Performance in Different Subjects")
plt.ylabel("Average Score")
plt.show()
```

# Area PLot

```python
plt.fill_between(avg_scores.index, avg_scores.values,color='red',alpha=0.5)
plt.title("Area Plot:Average Scores Across Subjects")
plt.show()
```

# Scatter Plots with Regression Lines

```python
# Scatter plots helps us see relationships between two numeric values
# For example,we can check if students who score high in math also high in reading
```

```python
sns.lmplot(x='math score', y='reading score', data=df)
plt.title("Math vs Reading Scores with Regression Line")
plt.show()
```

# Distribution Plots(Histogram,Boxplot, Violin)

```python
sns.histplot(df['math score'], kde=True, color='blue')
plt.title("Distribution of Math Scores")
plt.show()
```

# Boxplot

```python
sns.boxplot(x='gender',y='math score',data=df)
plt.title("Math Score Distribution by Gender")
plt.show()
```

# Violin Plot

```python
sns.violinplot(x='gender',y='reading score', data=df)
plt.title("Reading Score Distribution by Gender")
plt.show()
```

# Correlation and Heatmaps



```python
#Correlation shows how strongly two variables are related
#A heatmap helps us see all correlations at once.Brighter colors=stronger relationships.
```

```python
corr = df[['math score','reading score','writing score']].corr(method='pearson')
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Scores')
plt.show()
```

# Combining Plots and Figure Styles

```python
fig,axes = plt.subplots(1,3, figsize=(15,5))

sns.histplot(df['math score'], ax=axes[0],color='skyblue')
sns.boxplot(y=df['reading score'],ax=axes[1], color='lightgreen')
sns.violinplot(y=df['writing score'], ax=axes[2], color='salmon')

plt.suptitle("Comparison of Score Distributions")
plt.show()

#Subplots make comparisons clearer and more organized.
```

# Pair Plots

```python
sns.pairplot(df[['math score','reading score','writing score']])
plt.suptitle("Pair Plot of Student Scores", y=1.02)
plt.show()

#Pair plots are great for multivariate EDA ,helping spot both correlations and outliers
```

# Categorical and Time-Series Visualization

```python
sns.barplot(x='parental level of education',y='math score',data=df)
plt.xticks(rotation=45)
plt.title("Math Score by Parental Education Level")
plt.show()
```

# Advanced Seaborn Plotting(Swarm and Violin Plots)

```python
#Combining violin and swarm plots gives us both distribution shape and individual data points
```

```python
sns.violinplot(x='gender', y='writing score', data=df, inner=None, color='lightblue')
sns.swarmplot(x='gender',y='writing score',data=df,color='black',size=3)
plt.title("Combined Violin and Swarm Plot:Writing Scores by Gender")
plt.show()
```

# Correlation Analysis(Pearson and Spearman)

```python
# Pearson correlation - Measures linear relationships
# Spearman correlation- measures rank(monotomic) relationships
# Both range between -1 and +1
```

# Step 1:Load the Dataset

```python
import pandas as pd, seaborn as sns, matplotlib.pyplot as plt

df = pd.read_csv('/kaggle/input/d/spscientist/students-performance-in-exams/StudentsPerformance.csv')
df.head()
```

# Step 2:Pearson Correlation

```python
#slect numeric data
numeric_df = df.select_dtypes(include=['int64', 'float64'])

pearson_corr = numeric_df.corr(method='pearson')
print("Pearson Correlation:\n",pearson_corr)
```

# Step 3:Spearman Correlation

```python
spearman_corr = numeric_df.corr(method='spearman')
print("Spearman Correlation:\n",spearman_corr)
```

# Step 4:Visualize with a heatmap

```python
plt.figure(figsize=(8,6))
sns.heatmap(pearson_corr, annot=True,cmap='coolwarm',fmt=".2f")
plt.title("Correlation Heatmap(Pearson)")
plt.show()
```
