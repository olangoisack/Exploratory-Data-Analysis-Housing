# Pandas DataFrames and Series

```python
#Series: one-dimensional labeled array(1D)
ages = pd.Series([22, 25, 30], name = "Age")
print("Series Example:\n",ages)
```

```python
#Dataframe;two-dimensional table(rows & columns)-2D
data = {
    "Name": ["Anne", "Bob", "Jepchirchir"],
    "Age" : [22, 25, 30],
    "City": ["Nairobi", "Lagos", "Ougadogou"]
}
data = pd.DataFrame(data)
print("\nDataFrame Example:\n",data)

```

# Indexing ,Filtering, and Slicing

```python
#Selecting columns
print(data["Name"]) #single column
```

```python
print(data[["Name","City"]]) #Multiple columns
```

```python
#Selecting rows
print(data.iloc[0]) #First row by index position
```

```python
print(data.loc[1, "City"]) #Row with index=1, column="City"
```

```python
#Filtering with conditions
```

```python
print(df_csv[df_csv["Score"] > 85])
```

```python
print(df_csv[(df_csv["Age"] > 23) & (df_csv["City"] == "Nairobi")])
```

```python
#Merging and Concatenation
#Concatenation Example
df1 = pd.DataFrame({"ID": [1,2], "Name": ["Alice","Bob"]})
df2 = pd.DataFrame({"ID": [1,2], "Score": [90, 85]})

concat_df = pd.concat([df1, df2], axis=1)
print("\nConcatenated DataFrame:\n", concat_df)
```
